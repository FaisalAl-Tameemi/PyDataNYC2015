```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
%matplotlib inline
```

```python
def beat_freq(f1, f2, f3, max_time, rate):
    t = np.linspace(0, max_time, rate*max_time)
    signal_1 = np.sin(2*np.pi*f1*t)
    signal_2 = np.sin(2*np.pi*f2*t)
    signal_3 = np.sin(2*np.pi*f3*t)
    display(Audio(data=signal_1, rate=rate))
    display(Audio(data=signal_2, rate=rate))
    display(Audio(data=signal_3, rate=rate))
    return signal_1, signal_2, signal_3
```

```python
signal_250, signal_275, signal_300 = beat_freq(250.0, 275.0, 300.0, 1, 11025)
```

```python
s = np.hstack((signal_275+signal_300, signal_275+signal_250))
```

```python
fig1 = plt.figure(figsize=(10, 8), facecolor='white')
plt.specgram(s, NFFT=4096, Fs=11025, noverlap=2048, cmap=plt.cm.gist_heat)
axes = plt.gca()
axes.set_ylim([0, 560])
plt.xlabel('Time [s]', fontsize=18)
plt.ylabel('Frequency [Hz]', fontsize=18)
plt.show()
```

```python
fig2 = plt.figure(figsize=(10, 8), facecolor='white')
plt.specgram(s, NFFT=1024, Fs=11025, noverlap=512, cmap=plt.cm.gist_heat)
axes = plt.gca()
axes.set_ylim([0, 600])
plt.xlabel('Time [s]', fontsize=18)
plt.ylabel('Frequency [Hz]', fontsize=18)
plt.show()
```
